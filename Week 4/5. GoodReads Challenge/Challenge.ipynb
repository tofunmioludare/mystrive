```python
import pandas as pd

df = pd.read_csv("Dataset.csv")
```

```python
df.sample(10)
```

```python
df.dtypes
```

```python
df["minmax_norm_ratings"] = (1 + ((df["avg_rating"]-df["avg_rating"].min())/(df["avg_rating"].max()-df["avg_rating"].min()))*9).round(2)
df.sample(10)
```

```python
df.to_csv("groupChallenge.csv", index = True)
```

```python
df["mean_norm_ratings"] = (1 + ((df["avg_rating"]-df["avg_rating"].mean())/(df["avg_rating"].max()-df["avg_rating"].min()))*9).round(2)
df.sample(10)
```

```python
df.to_csv("groupChallenge.csv", index = True)
```

```python
publishYearGroup = df.groupby("original_publish_year")
publishYearGroup.first()
```

```python
publishYearGroupMean = publishYearGroup.agg({"minmax_norm_ratings": ["mean"]}).round(2)
print(publishYearGroupMean)
```

```python
def getBestBook(authorName):
    result = df.groupby("author").get_group(authorName).sort_values("minmax_norm_ratings", ascending=False).iloc[0]
    return result
print(getBestBook("Jerry Spinelli"))
```
